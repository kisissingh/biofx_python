```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_batch(batch_id, condition="Good"):
    # Time: 100 hours, measurements every 1 hour
    time = np.linspace(0, 100, 100)

    # 1. BIOMASS (The Cells): Logistic Growth (S-Curve)
    # Cells grow slowly, then fast, then stop.
    carrying_capacity = 10.0 # Max density
    growth_rate = 0.1
    biomass = carrying_capacity / (1 + np.exp(-growth_rate * (time - 50)))

    # 2. DISSOLVED OXYGEN (DO): Inverse to Biomass
    # As cells grow, they consume Oxygen.
    do_level = 100 - (biomass * 8) # Simple inverse relationship

    # 3. ADDING REALISM (Noise)
    # Sensors are never perfect. We add random "jitter".
    noise_level = 0.5 if condition == "Good" else 2.0
    biomass += np.random.normal(0, 0.1, 100)
    do_level += np.random.normal(0, noise_level, 100)

    # 4. SIMULATING A FAILURE (The "Bad Batch")
    # If bad, the Oxygen sensor fails at hour 60.
    if condition == "Bad":
        do_level[60:] = 0 # Sensor dies or Oxygen cut off

    # Pack it into a DataFrame
    df = pd.DataFrame({
        "Time": time,
        "Batch_ID": batch_id,
        "Biomass_gL": biomass,
        "Dissolved_Oxygen_pct": do_level,
        "Condition": condition
    })
    return df

# --- EXECUTION ---
# Generate 50 Good Batches (The "Golden" History)
history_list = []
for i in range(50):
    history_list.append(generate_batch(batch_id=f"BATCH_G_{i}", condition="Good"))

# Generate 5 Bad Batches (For testing failures)
for i in range(5):
    history_list.append(generate_batch(batch_id=f"BATCH_B_{i}", condition="Bad"))

# Combine and Save
full_data = pd.concat(history_list)
full_data.to_csv("bioreactor_data.csv", index=False)

print("‚úÖ Factory Simulation Complete. Data saved to 'bioreactor_data.csv'")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# 1. LOAD DATA
# We read the file we just made.
df = pd.read_csv("bioreactor_data.csv")

# 2. PREPARE TRAINING DATA (The "Golden" Standard)
# We only train on 'Good' batches. The AI should not know what 'Bad' looks like yet.
good_data = df[df['Condition'] == 'Good']
features = ['Biomass_gL', 'Dissolved_Oxygen_pct']

# Standardize the data (Scale it so Mean=0, Variance=1)
# This is crucial for PCA.
scaler = StandardScaler()
X_train = scaler.fit_transform(good_data[features])

# 3. TRAIN THE BRAIN (PCA)
# We compress 2 variables into 1 "Principal Component".
# If the data follows the rules, this compression is perfect.
# If the data breaks the rules, the compression fails (Error).
pca = PCA(n_components=1)
pca.fit(X_train)

# 4. DEFINE THE "ALARM" LOGIC (T-Squared / Reconstruction Error)
def get_anomaly_scores(data, pca_model, scaler_model):
    # Scale new data
    X_scaled = scaler_model.transform(data[features])

    # Project into PCA space (The "Golden Tunnel")
    X_pca = pca_model.transform(X_scaled)

    # Project back to original space
    X_reconstructed = pca_model.inverse_transform(X_pca)

    # Calculate Error (Distance between Original and Reconstructed)
    # High Error = The data broke the correlation rules.
    reconstruction_error = np.sum(np.square(X_scaled - X_reconstructed), axis=1)
    return reconstruction_error

# 5. TEST IT: Let's run a "Bad Batch" through the Brain
bad_batch = df[df['Batch_ID'] == 'BATCH_B_0'] # One of the failures
bad_batch_scores = get_anomaly_scores(bad_batch, pca, scaler)

# 6. VISUALIZE THE RESULT
plt.figure(figsize=(12, 5))

# Plot 1: The Sensor Data (What the Human sees)
plt.subplot(1, 2, 1)
plt.plot(bad_batch['Time'], bad_batch['Dissolved_Oxygen_pct'], color='red', label='Oxygen (Sensor)')
plt.axvline(x=60, color='black', linestyle='--', label='Sensor Failure Event')
plt.title("What the Operator Sees (Raw Data)")
plt.legend()

# Plot 2: The AI Alarm (What the Digital Twin sees)
plt.subplot(1, 2, 2)
plt.plot(bad_batch['Time'], bad_batch_scores, color='purple', linewidth=2, label='Anomaly Score')
plt.axhline(y=3.0, color='red', linestyle='--', label='Alarm Threshold') # Simple threshold
plt.title("What the AI Sees (Anomaly Score)")
plt.legend()

plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. SETUP THE DASHBOARD LOGIC
def update_dashboard(current_hour):
    # Get data for this specific moment
    # Use == current_hour to get the exact row for the current time
    row = bad_batch[bad_batch['Time'] == current_hour].iloc[0]
    # Use current_hour directly as the index for bad_batch_scores
    score = bad_batch_scores[int(current_hour)]

    # Determine Status
    status = "‚úÖ NORMAL"
    color = "green"

    if score > 3.0: # The Threshold
        status = "üö® CRITICAL FAILURE DETECTED"
        color = "red"

    # 2. DRAW THE "Screen"
    print(f"‚è±Ô∏è BATCH TIME: {current_hour:.1f} Hours")
    print("-" * 30)
    print(f"üìâ DISSOLVED OXYGEN: {row['Dissolved_Oxygen_pct']:.2f}%")
    print(f"üß† AI ANOMALY SCORE: {score:.2f}")
    print("-" * 30)

    # The Big Alert Banner
    # We use ANSI codes to make it bold/colored in the output
    if color == "red":
        print(f"\033[1;31;40m {status} \033[0m") # Red Text
        print("ACTION: PAUSING FEED PUMP...")
    else:
        print(f"\033[1;32;40m {status} \033[0m") # Green Text

# 3. CREATE THE SLIDER
slider = widgets.FloatSlider(
    value=0,
    min=0,
    max=99,
    step=1,
    description='Time (Hrs):',
    continuous_update=True
)

# 4. LINK SLIDER TO DASHBOARD
ui = widgets.interactive(update_dashboard, current_hour=slider)
display(ui)
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# --- PAGE CONFIGURATION ---
st.set_page_config(page_title="BioTwin: Golden Batch System", layout="wide")

st.title("üß¨ BioTwin: Real-Time Release (RTR) Dashboard")
st.markdown("**Status:** FDA 21 CFR Part 11 Compliant | **System:** Bioreactor B-204")

# --- 1. THE "FAKE FACTORY" (Data Gen) ---
@st.cache_data # Caches the data so it doesn't reload every time
def generate_data():
    time_steps = np.linspace(0, 100, 100)

    # Generate 50 Good Batches (Training Data)
    good_batches = []
    for i in range(50):
        biomass = 10.0 / (1 + np.exp(-0.1 * (time_steps - 50))) + np.random.normal(0, 0.1, 100)
        do_level = 100 - (biomass * 8) + np.random.normal(0, 0.5, 100)
        good_batches.append(np.column_stack((biomass, do_level)))

    # Generate 1 Bad Batch (Sensor Failure at 60h)
    bad_biomass = 10.0 / (1 + np.exp(-0.1 * (time_steps - 50))) + np.random.normal(0, 0.1, 100)
    bad_do = 100 - (bad_biomass * 8) + np.random.normal(0, 0.5, 100)
    bad_do[60:] = 0 # FAILURE
    bad_batch = pd.DataFrame({'Time': time_steps, 'Biomass': bad_biomass, 'DO': bad_do})

    return np.vstack(good_batches), bad_batch

# Load Data
X_train, bad_batch = generate_data()

# --- 2. THE "BRAIN" (Model Training) ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

pca = PCA(n_components=1)
pca.fit(X_train_scaled)

# Calculate Anomaly Scores for the Bad Batch
X_bad_scaled = scaler.transform(bad_batch[['Biomass', 'DO']])
X_bad_pca = pca.transform(X_bad_scaled)
X_bad_recon = pca.inverse_transform(X_bad_pca)
recon_error = np.sum(np.square(X_bad_scaled - X_bad_recon), axis=1)

# --- 3. THE "CONTROL TOWER" (Dashboard Interface) ---

# Sidebar Controls
st.sidebar.header("üïπÔ∏è Simulation Controls")
current_time = st.sidebar.slider("Batch Time (Hours)", 0, 99, 0)
show_golden = st.sidebar.checkbox("Show Golden Tunnel", value=True)

# Get Data for Current Time
current_row = bad_batch.iloc[current_time]
current_score = recon_error[current_time]

# Top Metrics Row
kpi1, kpi2, kpi3 = st.columns(3)
kpi1.metric("Dissolved Oxygen", f"{current_row['DO']:.2f}%", delta="-0.5%" if current_time < 60 else "-100%")
kpi2.metric("Biomass Density", f"{current_row['Biomass']:.2f} g/L")
kpi3.metric("Anomaly Score", f"{current_score:.2f}", delta_color="inverse")

# Main Charts
col1, col2 = st.columns(2)

with col1:
    st.subheader("üì° Live Sensor Data (Dissolved Oxygen)")
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(bad_batch['Time'], bad_batch['DO'], color='red', label='Current Batch')
    if show_golden:
        ax.fill_between(bad_batch['Time'], 80, 110, color='gray', alpha=0.2, label='Golden Tunnel')
    ax.axvline(current_time, color='blue', linestyle='--', label='Current Time')
    ax.set_ylim(-10, 120)
    ax.legend()
    st.pyplot(fig)

with col2:
    st.subheader("üß† Digital Twin Diagnostics")
    fig2, ax2 = plt.subplots(figsize=(6, 4))
    # Color logic for the line
    ax2.plot(bad_batch['Time'], recon_error, color='purple', label='Deviation Score')
    ax2.axhline(3.0, color='red', linestyle='--', label='Critical Threshold')
    ax2.axvline(current_time, color='blue', linestyle='--')

    # Dynamic Status Message
    if current_score > 3.0:
        st.error(f"üö® CRITICAL FAILURE DETECTED at Hour {current_time}")
    else:
        st.success("‚úÖ Process Within Control Limits")

    st.pyplot(fig2)

# Footer
st.caption("BioTwin v1.0 | Powered by Python & Scikit-Learn")
```
